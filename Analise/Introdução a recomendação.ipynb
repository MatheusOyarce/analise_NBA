```python
import pandas as pd
import numpy as np
```

```python
# Importando os filmes e renomeando as colunas
filmes = pd.read_csv ('movies.csv')
filmes.columns = ['filmeId','titulo','genero']
filmes = filmes.set_index ('filmeId')
filmes.head()
```

```python
# Importando as notas
notas = pd.read_csv ('ratings.csv')
notas.columns = ['usuarioId','filmeId','nota','momento']
notas.head()
```

## Primeira tentativa de recomendação

```python
total_notas = notas['filmeId'].value_counts()
total_notas.head(10)
```

```python
filmes.loc[356]
```

```python
# Criando uma coluna nova em filmes que contera os dados de total de notas
filmes ['total_de_notas'] = total_notas
filmes
```

```python
# Vendo os top 10 filmes mais votados seguindo a coluna total de notas
filmes.sort_values('total_de_notas', ascending = False).head(10)
```

```python
# Média de votos 
notas_media = notas.groupby ('filmeId').mean() ['nota']
notas_media.head()
```

```python
# Criando coluna media de votos e atribuindo
# Nem sempre o mais avaliado, é o mais bem avaliado

filmes ['nota_media'] = notas_media
filmes.sort_values('total_de_notas', ascending = False).head(10)
```

```python
filmes_mais_20_votos = filmes.query ('total_de_notas >= 20')
filmes_mais_20_votos.sort_values('nota_media', ascending = False).head(10)
```

## Comparando Usuarios

1. Pegando notas por usuarioId;
2. Extraindo notas de cada usuario atrávez de funções

```python
# Buscando as notas por usuario
def notas_do_usuario(usuario):
  notas_do_usuario = notas.query("usuarioId==%d" % usuario)
  notas_do_usuario = notas_do_usuario[["filmeId", "nota"]].set_index("filmeId")
  return notas_do_usuario

# Função para definir a distancia entre usuarios
def distancia_vetores (a,b):
    return np.linalg.norm(a - b)
```

```python
usuario1 = notas_do_usuario (1)
usuario4 = notas_do_usuario (4)
```

```python
# Juntas os dois DataFrame
# por terem a mesma coluna chamada nota, temos que fazer um suffix para diferenciar cada coluna
# Aqui ja estamos juntando as duas colunas e excluindo os valores nulos
diferenca = usuario1.join(usuario4, lsuffix = '_esquerda', rsuffix = '_direita').dropna()
distancia_vetores(diferenca['nota_esquerda'],diferenca['nota_direita'])
```

O .join com base no indice vai usar os dados do usuario1 como base para buscar os filmes similares aos dele no conjunto do usuario2. Os filmes que ele não achar, vai voltar com resultado nan, então com o drop ja excluimos e deixamos apenas os filmes similares.


```python
# Criando função para calcular a distancia entre usuarios de qualquer id dentro do conjunto
def distancia_de_usuario (usuario_id1, usuario_id2):
    notas1 = notas_do_usuario (usuario_id1)
    notas2 = notas_do_usuario (usuario_id2)
    diferenca = notas1.join(notas2, lsuffix = '_esquerda', rsuffix = '_direita').dropna()
    return distancia_vetores (diferenca['nota_esquerda'],diferenca['nota_direita'])
```
